Load the data in 'world_bank_turkey.csv' file.

Change the labels as follows (old label --> new label).

Urban population (% of total population) --> Urban Ratio

Population, total --> Total

Population, male -->Male

Population, female --> Female

Population growth (annual %) --> Population Growth

Life expectancy at birth, total (years) --> Life Expectancy

Armed forces personnel, total --> Armed Forces Personnel

Literacy rate, adult total (% of people ages 15 and above) --> Literacy Rate


```python
import pandas as pd
world_bank_turkey = pd.read_csv('world_bank_turkey.csv')
world_bank_turkey.rename(columns = {"Urban population (% of total population)":"Urban Ratio" , "Population, total" : "Total" , "Population, male":"Male" , "Population, female" : "Female" , "Population growth (annual %)":"Population Growth" , "Life expectancy at birth, total (years)" : "Life Expectancy", "Armed forces personnel, total" : "Armed Forces Personnel", "Literacy rate, adult total (% of people ages 15 and above)" : "Literacy Rate"}, inplace = True)
world_bank_turkey
```

```python
# Show the first 15 rows of the table
world_bank_turkey.head(15)
```

```python
# Add a new column to the table which shows the Rural ratio of the population. Name the column as 'Rural Ratio'
rural_ratio = world_bank_turkey["Rural population"] / world_bank_turkey["Total"]
world_bank_turkey.insert(4, "Rural Ratio", rural_ratio*100) # i prefered to insert it as the fifth column
world_bank_turkey.head(5) # let's check
```
