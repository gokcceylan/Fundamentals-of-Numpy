**Read the data in "text.txt" file uploaded on piazza (You might need google how to read files in python). Split the text by space character. Then visualize the distribution of word counts. That is, we want to know whether words are used in a balanced way or there are some words used much frequently than others.**

```python
#write your answer here
# This code is part of one of my Coursera capstone projects. I didn't get it from another student or from stack overflow

!pip install fileupload
!pip install ipywidgets
!jupyter nbextension install --py --user fileupload
!jupyter nbextension enable --py fileupload

from matplotlib import pyplot as plt
from IPython.display import display
import fileupload
import io
import sys

def _upload():

    _upload_widget = fileupload.FileUploadWidget()

    def _cb(change):
        global file_contents
        decoded = io.StringIO(change['owner'].data.decode('utf-8'))
        filename = change['owner'].filename
        print('Uploaded `{}` ({:.2f} kB)'.format(
            filename, len(decoded.read()) / 2 **10))
        file_contents = decoded.getvalue()

    _upload_widget.observe(_cb, names='data')
    display(_upload_widget)

_upload()
```

```python
def calculator_of_words(file_contents):
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    uninteresting_words = ["the", "a", "to", "if", "is", "it", "of", "and", "or", "an", "as", "i", "me", "my", \
    "we", "our", "ours", "you", "your", "yours", "he", "she", "him", "his", "her", "hers", "its", "they", "them", \
    "their", "what", "which", "who", "whom", "this", "that", "am", "are", "was", "were", "be", "been", "being", \
    "have", "has", "had", "do", "does", "did", "but", "at", "by", "with", "from", "here", "when", "where", "how", \
    "all", "any", "both", "each", "few", "more", "some", "such", "no", "nor", "too", "very", "can", "will", "just"]
    #i noticed that there were a lot of interesting words that were frequently used such as "of" or "a" so i googled
    #a string of uninteresting words. i also noticed that "However," was seen as a different word than "however" so
    #i casefolded the string and then eliminated the punctuations before counting the words :) - Gokce
    words_in_file = file_contents.split()
    frequencies = {}
    
    for word in words_in_file:
        if word in uninteresting_words:
            pass
        else:
            for letter in word:
                if letter in punctuations:
                    letter.replace(punctuations,"")
            if word not in frequencies.keys():
                frequencies[word]=0
            frequencies[word]+=1
    return frequencies

file_contents = file_contents.casefold()
```

```python
temp_dict = calculator_of_words(file_contents)
temp_values = temp_dict.values()
temp_keys = temp_dict.keys()
temp_array = [temp_keys, temp_values]
freq_dataframe = pd.DataFrame(temp_array)
freq_dataframe = freq_dataframe.transpose()
freq_dataframe.rename(columns = {0:"words", 1:"frequencies"}, inplace = True)
freq_dataframe.sort_values("frequencies").plot.barh("words", "frequencies", figsize = [5,200])
```
