# Dataframes
Pandas uses dataframes to store data instead of arrays or numpy arrays. You can think of dataframes like tables. There are a bunch of cool stuff you can do with pandas and dataframes

```python
import pandas as pd
```

```python
data = {'apple' : [5, 10, 15, 20],
       'orange' : [5, 3, 15, 20],
       'banana' : [5, 10, 15, 20]}
data
```

```python
# turn that dictionary to a dataframe
our_dataframe = pd.DataFrame(data)
our_dataframe
```

```python
our_dataframe.head(0) #gives the column names
```

```python
our_dataframe.head(5) #gives the first 5 rows
```

```python
#create a dataframe with the given index
our_dataframe = pd.DataFrame(data, index =["Ahmet", "Ali", "Ayşe", "Veli"])
```

```python
our_dataframe #indexes are changed
```

# Note
As you can see above, when you set a column as an index, you can't get it like how you would get any other column

```python
#if we want that indexed player column, we should reset our index
happiness2015_indexed = happiness2015_indexed.reset_index()
happiness2015_indexed #we destroyed our index
#be careful
```

```python
#we can filter the non-wanted columns
happiness2015.drop(columns={"Generosity", "Dystopia Residual"})
```

```python
# what we do here is like SQL WHERE query
# get particular *rows* using loc
happiness2015.loc[0:3]
```

```python
#iloc also works with a slight difference
happiness2015.iloc[0:3] #not inclusive just like range() function
```

```python
#there's a bigger difference though :)
happiness2015_indexed = pd.read_csv('2015csv.csv', index_col=0)
happiness2015_indexed
```

loc locates by name but iloc locates by numerical index

```python
#let's get the countries between Canada and Turkey
happiness2015_indexed.loc["Canada":"Turkey"]
```

```python
#let's use iloc with the data frames indexed by country
happiness2015_indexed.iloc["Canada": "Turkey"]
```

```python
happiness2015_indexed.iloc[4:76]
```

* non-numeric indexes can be used with loc
* if you want to choose with index, use iloc
* if you want to choose by name, use loc

```python
#get elements at specific indices using numpy arange
happiness2015_indexed.iloc[4:76:5]
```

# Sorting

```python
#sort by GDP
happiness2015.sort_values("GDP")
```

```python
#sort by GDP in descending order
happiness2015.sort_values("GDP", ascending=False)
```

```python
#get countries whose GDP is more than 1.5
happiness2015[happiness2015["GDP"] > 1.5] #returns dataframe
```

```python
happiness2015["GDP"] > 1.5 #returns array
```

```python
#get Western Eu countries
happiness2015[happiness2015["Region"] == "Western Europe"]
```

```python
# get Western EU countries with a happiness score of over 6.5
happiness2015[(happiness2015["Region"] == "Western Europe") and (happiness2015["Happiness Score"] > 6.5)]
# and fail trying...
```

```python
#get Western EU countries with a happiness score of over 6.5
happiness2015[(happiness2015["Region"] == "Western Europe") & (happiness2015["Happiness Score"] > 6.5)]
# this is the right way :)
```

```python
#get countries that are either in Europe or have a gdp above 1.5
happiness2015[(happiness2015["Region"].str.contains("Europe")) | (happiness2015["GDP"] > 1.5)]
```

```python
#countries who are in a region that starts with "Western"
happiness2015[(happiness2015["Region"].str.startswith("North"))] 
```
